```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from pathlib import Path
```

```python
csv_file = Path('CSV Files/CIB')
hours = np.arange(24)
```

```python
sns.set_theme(
    style='darkgrid',
    rc={
        'axes.edgecolor': 'black',
        'axes.linewidth': 1,
        'xtick.color': 'black',
        'ytick.color': 'black',
        'xtick.major.width': 1,
        'ytick.major.width': 1,
        'xtick.major.size': 6,
        'ytick.major.size': 6,
        'xtick.bottom': True,
        'ytick.left': True
    }
)
```

```python
df_turner = pd.read_csv(csv_file / "aranet_turner_feb.csv")
df_turner['Time'] = pd.to_datetime(df_turner['Time'], dayfirst=True)
df_turner = df_turner.set_index('Time', drop=False)
```

```python
df_turner['hour'] = df_turner['Time'].dt.hour
df_turner['weekend'] = df_turner['Time'].dt.dayofweek >= 5

weekday_co2 = df_turner[~df_turner['weekend']].groupby('hour')['CO₂, ppm'].mean()
weekend_co2 = df_turner[df_turner['weekend']].groupby('hour')['CO₂, ppm'].mean()

plt.figure(figsize=(12, 6))
plt.plot(df_turner.index, df_turner['CO₂, ppm'], alpha=0.3, label='Raw $CO_2$ Concentration')
plt.plot(df_turner.index, df_turner['CO₂, ppm'].rolling(12).mean(), linewidth=2, label='2 hour rolling mean')
plt.title('Turner Lab $CO_2$ Concentration Over Time (February)', fontsize=18)
plt.xlabel('Time', fontsize=16)
plt.ylabel('$CO_2$ Concentration (ppm)', fontsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.legend()
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=4))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
plt.tight_layout()

plt.figure(figsize=(12, 6))
sns.lineplot(x=hours, y=weekday_co2, label='Weekdays', marker='.')
sns.lineplot(x=hours, y=weekend_co2, label='Weekends', marker='.')
plt.title('Average $CO_2$ Levels by Hour (Weekdays vs Weekends in February)', fontsize=18)
plt.xlabel('Hour of the Day', fontsize=16)
plt.ylabel('Average $CO_2$ (ppm)', fontsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.xticks(hours)
plt.legend()
#plt.xlim(0, 23)
plt.tight_layout()
```

```python
weekday_temp = df_turner[~df_turner['weekend']].groupby('hour')['Temperature, °C'].mean()
weekend_temp = df_turner[df_turner['weekend']].groupby('hour')['Temperature, °C'].mean()

plt.figure(figsize=(12, 6))
plt.plot(df_turner.index, df_turner['Temperature, °C'], alpha=0.3, label='Raw Temperature')
plt.plot(df_turner.index, df_turner['Temperature, °C'].rolling(12).mean(), linewidth=2, label='2 hour rolling mean')
plt.title('Turner Lab Temperature Over Time (February)', fontsize=18)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Temperature (°C)', fontsize=16)
plt.legend()
plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=4))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
plt.tight_layout()

plt.figure(figsize=(12, 6))
sns.lineplot(x=hours, y=weekday_temp, label='Weekdays', marker='.')
sns.lineplot(x=hours, y=weekend_temp, label='Weekends', marker='.')
plt.title('Average Temperature by Hour (Weekdays vs Weekends in February)', fontsize=18)
plt.xlabel('Hour of the Day', fontsize=16)
plt.ylabel('Average Temperature (°C)', fontsize=16)
plt.tick_params(axis='x', labelsize=16)
plt.tick_params(axis='y', labelsize=16)
plt.xticks(hours)
plt.legend()
#plt.xlim(0, 23)
plt.tight_layout()
```
